```r
# Load required libraries
library(parallel)
library(foreach)
library(doParallel)

# Create a list of 1000 larger random matrices
matrices <- replicate(1000,
                      matrix(rnorm(10000),  # Increased matrix size
                             ncol=1000),
                      simplify=FALSE)


# Define a function to compute the sum of the elements in a matrix
sum_matrix <- function(mat) {
  sum(mat)
}

# Compute the sums of the matrices using foreach with 4 cores
cl <- makeCluster(4)
clusterExport(cl, "matrices")
start_time <- Sys.time()
sums <- foreach(mat = matrices) %dopar% sum_matrix(mat)
end_time <- Sys.time()
stopCluster(cl)

# Compute the sums of the matrices using a for loop (serial computation)
start_time_serial <- Sys.time()
sums_serial <- numeric(length(matrices))
for (i in seq_along(matrices)) {
  sums_serial[i] <- sum_matrix(matrices[[i]])
}
end_time_serial <- Sys.time()

# Print the execution times
cat("Parallel execution time:", end_time - start_time, "\n")
cat("Serial execution time:", end_time_serial - start_time_serial, "\n")
```

```r
library(foreach)
library(doParallel)

# Create a list of 1000 random vectors
vectors <- replicate(1000, rnorm(1000),
                     simplify = FALSE)

# Define a function to compute the mean of a vector
mean_vector <- function(vec) {
  mean(vec)
}

# Compute the means of the vectors
# using foreach with 4 cores
cl <- makeCluster(4)
clusterExport(cl,"vectors")
start_time <- Sys.time()
means <- foreach(vec = vectors) %dopar% mean_vector(vec)
end_time <- Sys.time()
stopCluster(cl)

# Compute the means of the vectors using a for loop
start_time_serial <- Sys.time()
means_serial <- numeric(length(vectors))
for (i in seq_along(vectors)) {
  means_serial[i] <- mean_vector(vectors[[i]])
}
end_time_serial <- Sys.time()

# Print the execution times
cat("Parallel execution time:",
    end_time - start_time, "\n")
cat("Serial execution time:",
    end_time_serial - start_time_serial, "\n")
```

# Regression trees

```r
# Set the seed for reproducibility
set.seed(123)

# Define the number of observations to generate
n <- 1000

# Generate three predictor variables from a uniform distribution
x1 <- runif(n)
x2 <- runif(n)
x3 <- runif(n)

# Generate the response variable as a function of the predictor variables
y <- 2*x1 + 3*x2^2 + 0.5*sin(2*pi*x3) + rnorm(n, mean = 0, sd = 1)

# Create a data frame containing the predictor variables and the response variable
sim_data <- data.frame(x1 = x1, x2 = x2, x3 = x3, y = y)
```

```r
# Scatter plot of x1 vs y with linear regression line (without formula)
ggplot(sim_data, aes(x = x1, y = y)) +
  geom_point() +
  geom_smooth(se = FALSE)

# Scatter plot of x2 vs y with quadratic regression line
ggplot(sim_data, aes(x = x2, y = y)) +
  geom_point() +
  geom_smooth(se = FALSE)

# Scatter plot of x3 vs y with sinusoidal regression line
ggplot(sim_data, aes(x = x3, y = y)) +
  geom_point() +
  geom_smooth(se = FALSE)
```

```r
# Load the rpart package
library(rpart)

# Build the regression tree
tree <- rpart(y ~ x1 + x2 + x3,
              data = sim_data,
              control = rpart.control(minsplit = 10, cp = 0.01))
```

```r
# Plot the tree
plot(tree)
text(tree)
```

```r
# New data for prediction
new_data <- data.frame(x1 = 0.5, x2 = 0.5, x3 = 0.5)

# Predict the outcome for new data
y1=predict(tree, new_data)
cat(y1)
```

# Standardised Residuals

```r
# Use the built-in mtcars dataset
model <- lm(mpg ~ disp + hp, data = mtcars)

# View a brief summary of the model
summary(model)
```

```r
# Calculate internally studentized (standardized) residuals
std_resids <- rstandard(model)

# Show the first six values
print(head(std_resids))
```

```r
library(ggplot2)

df_plot <- data.frame(
  fitted = fitted(model),
  std_resid = std_resids
)

ggplot(df_plot, aes(x = fitted, y = std_resid)) +
  geom_point() +
  geom_hline(yintercept = c(-2, 2), linetype = "dashed") +
  labs(
    x = "Fitted Values",
    y = "Standardized Residuals",
    title = "Residuals vs. Fitted"
  )
```

```r
ggplot(df_plot, aes(x = std_resid)) +
  geom_histogram(binwidth = 0.5) +
  labs(
    x = "Standardized Residual",
    y = "Count",
    title = "Histogram of Standardized Residuals"
  )
```

```r
ggplot(df_plot, aes(sample = std_resid)) +
  stat_qq() +
  stat_qq_line() +
  labs(
    title = "Q–Q Plot of Standardized Residuals"
  )
```
